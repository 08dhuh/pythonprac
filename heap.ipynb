```python
import math

class MinHeap:
    def __init__(self,):
        self.heap_list = [None]
        self.count = 0
    
    def __str__(self):
        return self.print_list()

    def add(self, element):
        #print(f'“Adding {element} to {self.heap_list}.”')
        self.count += 1
        self.heap_list.append(element)
        self.heapify_up()
    
    def heapify_up(self):
        #print(f'“Restoring the heap property…”')
        idx = len(self.heap_list) - 1

        while self.parent_idx(idx) > 0:
            parent_idx = self.parent_idx(idx)
            child = self.heap_list[idx]
            parent = self.heap_list[parent_idx]
            if parent > child:
                #print(f'“swapping {parent}_element with element_at_{idx}“')
                self.heap_list[idx] = parent
                self.heap_list[parent_idx] = child
            idx = parent_idx

    def heapify_down(self):
        index = 1
        while self.left_child_idx(index) <= self.count:
            #print(f'current: {index}')
            lch_i = self.left_child_idx(index)            
            rch_i = self.right_child_idx(index)
            parent = self.heap_list[index]
            swap_index = lch_i if rch_i > self.count or self.heap_list[lch_i] <= self.heap_list[rch_i] else rch_i
            swap_property = self.heap_list[swap_index]
            
            if swap_property < parent:
                self.heap_list[swap_index] = parent
                self.heap_list[index] = swap_property
            index = swap_index         

    def parent_idx(self, idx):
        return idx // 2

    def left_child_idx(self, idx):
        return idx * 2

    def right_child_idx(self, idx):
        return idx * 2 + 1
    
    def retrieve_min(self):
        if self.count == 0:
            print('No items in heap')
            return None
        minimum = self.heap_list[1]
        print(f'Removing: {minimum} from {self.heap_list}')
        self.heap_list[1] = self.heap_list[-1]
        self.heap_list.pop()
        self.count -= 1
        print(f'“Last element moved to first: {self.heap_list}“')
        self.heapify_down()

    def print_list(self):
        string=''
        for i, item in enumerate(self.heap_list):
            if i==0: continue
            string += f'{item} '
            if math.floor(math.log2(i)) < math.floor(math.log2(i + 1)):
                string += '\n'
        return string
    
    
```

```python
min_heap = MinHeap()

# add elements
min_heap.add(7)
min_heap.add(12)
min_heap.add(42)
# Adding more elements to the min heap
min_heap.add(3)
min_heap.add(17)
min_heap.add(8)
min_heap.add(25)
min_heap.add(1)
min_heap.heap_list = [None, 20, 62, 27, 76, 97, 74, 62, 79, 77]
print(min_heap)
# remove minimum element
print(min_heap.retrieve_min())

print(min_heap)
```
